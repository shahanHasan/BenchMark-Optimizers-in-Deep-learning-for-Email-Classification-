```python
Optimizer_list = ["Adam", "Nadam", "SGD", "RMSProp", "Adagrad", "Adadelta", "Ftrl", "SGD-Momentum", "SGD-Nesterov-Momentum", "Adam-Weight-Decay"] 
Model_Architectures = ["BRNN", "Bi-LSTM", "Bi-GRU"]
cols = ['Optimizer', 'Precision', 'Recall', 'f1 score', "Train Accuracy", 'Test Accuracy', 'Loss', 'ROC-AUC', 'Train Time (s)']
Averages_Enron = []
Averages_LingSpam = []
Averages_SpamAssasin = []

import re
for i in Optimizer_list:
    avg_enron = ENRON_ALL[ENRON_ALL.Model.str.contains(f'{i}$', regex= True, na=False)].mean().tolist()  
    avg_lingspam = LINGSPAM_ALL[LINGSPAM_ALL.Model.str.contains(f'{i}$', regex= True, na=False)].mean().tolist()
    avg_spamassasin = SPAMASSASIN_ALL[SPAMASSASIN_ALL.Model.str.contains(f'{i}$', regex= True, na=False)].mean().tolist()
    avg_enron.insert(0, i)
    avg_lingspam.insert(0, i)
    avg_spamassasin.insert(0, i)
    Averages_Enron.append(avg_enron)
    Averages_LingSpam.append(avg_lingspam)
    Averages_SpamAssasin.append(avg_spamassasin)
    
```

```python
path = [["Enron_optim", Averages_Enron], ["LingSpam_optim", Averages_LingSpam], ["SpamAssasin_optim", Averages_SpamAssasin]]
for i in path:
    df_metrics = pd.DataFrame(i[1])
    df_metrics.columns = cols
    df_metrics.sort_values(by = 'Test Accuracy', ascending = False, inplace=True)
    df_metrics.reset_index(drop = True, inplace=True)
    df_metrics.to_csv(f"Comparison/{i[0]}.csv", index = False)
```

```python
Enron_average_optim = pd.read_csv("Comparison/Enron_optim.csv")
LingSpam_average_optim = pd.read_csv("Comparison/LingSpam_optim.csv")
SpamAssasin_average_optim = pd.read_csv("Comparison/SpamAssasin_optim.csv")
```

```python
Overall_across_dataset_avg = []
for i in Optimizer_list:
    avg = df[df.Optimizer.str.contains(f'{i}$', regex= True, na=False)].mean().tolist()
    avg.insert(0, i)
    Overall_across_dataset_avg.append(avg)
```

```python
df_metrics = pd.DataFrame(Overall_across_dataset_avg)
df_metrics.columns = cols
df_metrics.sort_values(by = 'Test Accuracy', ascending = False, inplace=True)
df_metrics.reset_index(drop = True, inplace=True)
df_metrics.to_csv(f"Comparison/Average_Overall_dataset.csv", index = False)
```
